```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
x=np.array([i for i in range(11)])
y=x**2
plt.plot(x,y,'rx-')
z=x**1.5
plt.plot(x,z,'go-')
z2=x**3.5
plt.plot(x,z2,'b<-')
plt.title("helloworld title!")
plt.legend(['1231'],loc=1)
plt.axis([0,10,0,120])
plt.text(2,80,'hello')
plt.grid(True)
plt.xlabel("this is x axis")
plt.ylabel("this is y axis")
plt.text(4,80,'$y = x^2$',bbox={'facecolor':'yellow','alpha':0.2})
plt.show()
# plt.savefig("a.png")
# print "ok"
```
