```python
import numpy as np
import matplotlib.pyplot as plt 
x=np.arange(-2.0*np.pi,2*np.pi,0.01)
for i in range(1,4):
    y=np.sin(float(i)*x)/x
    plt.plot(x,y)
plt.xticks(np.array([-2,-1,0,1,2])*np.pi,[-2,-1,0,1,2])
ax=plt.gca()
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_position(('data',0))
ax.spines['left'].set_color('none')
plt.show()
```
